Класс для векторизации текста различными предобученными моделями 

```python
class Text_comparison():
    def __init__(self,model_name):
        self.tokenizer = transformers.AutoTokenizer.from_pretrained(model_name)
        self.model = transformers.AutoModel.from_pretrained(model_name)
        
    def text_tokenizator(self, dataframe):
        result = []
        max_len = 0
        for title in dataframe:
            title = " ".join(title.split())
            inputs = self.tokenizer.encode_plus(
                title,
                None,
                add_special_tokens = True,
                return_token_type_ids = True
            )
            result.append(inputs.data)
            if len(inputs.data['input_ids']) > max_len:
                max_len = len(inputs.data['input_ids'])
        for i in result:
            i['input_ids'] = i['input_ids'] + [0]*(max_len - len(inputs.data['input_ids']))
            i['token_type_ids'] = i['token_type_ids'] + [0]*(max_len - len(inputs.data['token_type_ids']))
            i['attention_mask'] = i['attention_mask'] + [0]*(max_len - len(inputs.data['attention_mask']))
        return result
    
    def text_vectorization(self, dataframe):
        tokenized  = self.text_tokenizator(dataframe)
        
        input_ids = []
        attention_mask = []
        for i in tokenized:
            input_ids.append(i['input_ids'])
            attention_mask.append(i['attention_mask'])
        input_ids = torch.tensor(np.asarray(input_ids), dtype = torch.long)
        attention_mask = torch.tensor(np.asarray(attention_mask), dtype = torch.long)
        
        with torch.no_grad():
            last_hidden_states = self.model(input_ids, attention_mask = attention_mask)
            
        result = last_hidden_states[0].numpy()
        
        return result
    
    
    def text_vectorization_cls(self, dataframe):
        vectorized = self.text_vectorization(dataframe)
        result = vectorized[:,0,:]
        return result
```

```python
text_model = Text_comparison('DeepPavlov/rubert-base-cased')
```

Функция возвращает список всех имен изображений лежащих в папке

```python
def get_list_fail(directory):
    files = os.listdir(directory)
    l = []
    for f in files:
        l.append(int(f.split(".")[0]))
    return l
```

Фунция векторизирующая изображения в папке test

```python
def vectorize_test(n, model):
    path = "test/" + str(n) + ".png"
    img = load_img(path, target_size=(224, 224))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    pred = model.predict(x)
    return pred.ravel()
```

Фунция векторизирующая изображенияв папке train

```python
def vectorize_train(n, model):
    path = "train/" + str(n) + ".png"
    img = load_img(path, target_size=(224, 224))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    pred = model.predict(x)
    return pred.ravel()
```

Функция для чтения и векторизации описания и сообтетсвующеего данному экспонату изображения из train

```python
def get_train_text(fail_name):
    Train_text = []
    with open(f"{fail_name}.csv", 'r', encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            t = []
            t.append(text_model.text_vectorization_cls([row['description']]))
            t.append(row['object_img'])
            Train_text.append(t)
            print(len(Train_text), "/", 2098)
        return Train_text
```

Функция для чтения описания из test

```python
def get_test_text(fail_name):
    Test_text = []
    with open(f"{fail_name}.csv", 'r', encoding="utf-8") as csvfile:
        reader = csv.DictReader(csvfile)
        for row in reader:
            Test_text.append(row['description'])
    return Test_text
```

Получение названий и векторизизация изображений в test

```python
test_arr = get_list_fail("test")
v_arr = []
for image_numder in test_arr:
    v = vectorize_test(image_numder, image_model)
    print("Vectorized:", len(v_arr) + 1, "from", len(test_arr))
    v_arr.append([image_numder, v])
```
